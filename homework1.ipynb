```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import operator
```

```python
# 9.1
# Read the movie100.csv, and check average length (in terms of number of words) of the positive 
# and negative reviews, respectively

movies = pd.read_csv("review100.csv")
positives_reviews = movies.loc[movies["sentiment"] == "positive", ["movie review"]]
positives = positives_reviews["movie review"].str.split().str.len()
positive_avg_words = positives.mean()
print("Positive review average words count: ", positive_avg_words)

negatives_reviews = movies.loc[movies["sentiment"] == "negative", ["movie review"]]
negatives = negatives_reviews["movie review"].str.split().str.len()
negatives_avg_words = negatives.mean()
print("Negative review average words count: ", negatives_avg_words)
```

```python
# 9.2
# Check the frequency of the words with respect to positive and negative reviews, respectively. 
# Print the top 20 most frequent words (including words and their frequency), 
# for positive and negative reviews, respectively

positive_words = dict()
for index, row in positives_reviews.iterrows():
    for word in row["movie review"].split():
        if word not in positive_words:
            positive_words[word] = 1
        else:
            positive_words[word] = positive_words[word]+1
            
negative_words = dict()         
for index, row in negatives_reviews.iterrows():
    for word in row["movie review"].split():
        if word not in negative_words:
            negative_words[word] = 1
        else:
            negative_words[word] = negative_words[word]+1
            
positive_top_20 = dict(sorted(positive_words.items(), key=operator.itemgetter(1),reverse=True)[:20])
print("Top 20 words in Positive reviews:")
print(positive_top_20)

negative_top_20 = dict(sorted(negative_words.items(), key=operator.itemgetter(1),reverse=True)[:20])
print("\n\nTop 20 words in Negative reviews:")
print(negative_top_20)
```

```python
# 9.3
# Draw a bar plot to show the frequency of the two-20 most frequent words 
# (the x-axis denotes words, and the y-axis denotes their frequency), for positive and negative reviews, respectively

fig, axes = plt.subplots(1,2, figsize=(25, 15))


fig = plt.figure()
axes[0].bar(list(positive_top_20.keys()), list(positive_top_20.values()))
axes[0].set_title("Top 20 Positive review words")

axes[1].bar(list(negative_top_20.keys()), list(negative_top_20.values()), color='r')
axes[1].set_title("Top 20 Negative review words")

fig.tight_layout()
plt.show()
```
